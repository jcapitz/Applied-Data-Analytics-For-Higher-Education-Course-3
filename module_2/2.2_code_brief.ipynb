# 2.2 Code Brief: AI-Assisted Coding with Codex

Condensed reference for using Codex effectively.

## CRISP Prompting Framework

```
C — Context:      "I have a dataset of 5,000 students..."
R — Role:         "Act as an institutional research analyst..."
I — Instructions: "Build a logistic regression to predict..."
S — Specifics:    "Use L2 regularization, C=0.1, StandardScaler..."
P — Product:      "Show AUC, confusion matrix, top 5 coefficients"
```

## Quick Prompt Templates

**EDA:**
```
Show summary statistics, distributions, correlation heatmap,
and top features correlated with [TARGET].
```

**Classification:**
```
Build a [MODEL] with [PARAMS]. Show AUC, F1, precision, recall,
confusion matrix, and ROC curve.
```

**Comparison:**
```
Compare [MODEL1], [MODEL2], [MODEL3] in a table showing
AUC, F1, Precision, Recall. Plot all ROC curves together.
```
