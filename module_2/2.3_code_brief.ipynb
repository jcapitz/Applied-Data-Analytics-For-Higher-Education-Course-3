# 2.3 Code Brief: Choosing Your AI Coding Tool

Quick reference for Codex vs Colab + Gemini.

## Quick Decision

| Scenario | Use |
|:---------|:----|
| Quick exploration / prototype | **Codex** |
| Homework / coursework | **Colab + Gemini** |
| Formal report / reproducible analysis | **Colab + Gemini** |
| Learning new algorithms | **Colab + Gemini** |
| Fast iteration on model ideas | **Codex** |
| Team collaboration | **Colab + Gemini** |
| Large dataset with GPU needs | **Colab + Gemini** |

## Hybrid Workflow

```
1. EXPLORE with Codex  →  "What approach works best?"
2. BUILD in Colab      →  Implement in a reproducible notebook
3. SHARE the notebook  →  Google Drive / git
```

## Same CRISP Framework for Both

```
C — Context:      Describe your data
R — Role:         Specify your analytical perspective
I — Instructions: State the task clearly
S — Specifics:    Hyperparameters, features, methods
P — Product:      Desired outputs (charts, metrics, tables)
```
