# 1.3 Choosing Your AI Coding Tool
## Codex vs Google Colab + Gemini
- You've now seen both tools
- Both support vibecoding — different approaches
- This lesson helps you choose (or use both!)

## How They Work: Key Difference

**Codex** = CLI **agent in your terminal** that operates on your real project
```
TERMINAL  (in ~/projects/my-analysis)
┌────────────┐    ┌────────────────────┐
│ Your repo  │◄──►│ Codex agent        │
│ - data/    │    │ - reads files      │
│ - models/  │    │ - proposes diffs   │
│ - .git/    │    │ - runs commands    │
└────────────┘    └────────────────────┘
Edits saved to disk, tracked by git.
```

**Colab + Gemini** = AI assistant inside one cloud notebook
```
┌── COLAB NOTEBOOK ─────────────┐
│  Your code  ◄──► Gemini help  │
│  (you control every cell)     │
└───────────────────────────────┘
```

> **Codex spans your whole project. Gemini lives in one notebook.**

## Feature Comparison

| Feature | Codex | Colab + Gemini |
|:--------|:------|:---------------|
| Where it runs | Your terminal | Cloud notebook |
| What it edits | Any file in the repo | One notebook |
| Code visibility | Diff before each edit | Cells you write/inherit |
| Multi-file work | **Yes** | No |
| Persistence | On disk + git | Google Drive |
| GPU | Whatever your machine has | **Free T4** |
| Collaboration | Async via git / PRs | Real-time co-editing |

## Cost Comparison

| Tier | Codex | Colab + Gemini |
|:-----|:------|:---------------|
| Free | None — paid ChatGPT plan or API key required | Full Colab + basic Gemini |
| Paid | ChatGPT Plus $20/mo (or Pro / Team) | Colab Pro $12/mo |
| University | ChatGPT Edu / Enterprise | Google Workspace (often free) |

> For students: **Colab's free tier is more generous**

## Codex: Pros & Cons

### Advantages
- Edits **real files** in your **real project**
- Multi-file refactors and scaffolding
- Output is source code in git — fully reproducible
- Strong iteration loop with full project context

### Disadvantages
- Steeper setup (CLI, npm/Homebrew, auth)
- Need code literacy to review diffs
- Real consequences if you skip git hygiene
- Paid only — no meaningful free tier

## Colab + Gemini: Pros & Cons

### Advantages
- **Zero setup** — browser + Google account
- Generous free tier, free GPU
- Notebook-native (markdown + code + charts together)
- Easy sharing and real-time co-editing
- Great for **learning** — see every cell

### Disadvantages
- One notebook at a time — no multi-file work
- `.ipynb` files diff badly in git
- Session timeouts (~90 min idle)
- Less capable for complex agentic tasks

## When to Use Which?

```
Single notebook, one cloud file?
  → Colab + Gemini

Multi-file project, want git history?
  → Codex

No local Python / Chromebook?
  → Colab + Gemini

Need to refactor an existing repo?
  → Codex
```

## Scenario Guide for IR

| Scenario | Best Tool |
|:---------|:---------|
| Homework / coursework | Colab + Gemini |
| Quick exploration in one notebook | Colab + Gemini |
| Multi-file analysis repo | Codex |
| Refactoring an existing pipeline | Codex |
| Writing tests next to model code | Codex |
| Real-time pair work | Colab + Gemini |
| Async work via PR review | Codex |
| Learning a new algorithm | Colab + Gemini |

## The Hybrid Workflow: Use Both!

### Pattern 1: "Colab to Explore, Codex to Productionize"
1. **Explore** in Colab — sketch the approach in one notebook
2. **Promote** with Codex — turn it into a real repo (scripts, modules, tests)
3. **Share** the repo via git

### Pattern 2: "Codex Primary, Colab for Show-and-Tell"
1. **Work** in Codex day-to-day on the project
2. **Generate** a clean summary `.ipynb` with Codex
3. **Upload** to Colab to share with non-technical stakeholders

> Don't edit the **same file** in both at once — pick one as source of truth.

## Data Privacy Reminder

- **Both** tools send file/cell contents to a cloud model
- "Codex runs locally" doesn't mean your data stays local
- Check your institution's data governance for **both**
- FERPA compliance requires institutional review
- **For this course:** use the provided de-identified data

## The Bottom Line

| | Codex | Colab + Gemini |
|:--|:------|:---------------|
| Best for | Multi-file projects, refactors | Single-notebook work, learning |
| Where | Your terminal | Your browser |
| What it edits | The whole repo | One notebook |
| Cost | Paid only | Free tier |
| GPU | Local / remote box | Free T4 |

> **No wrong choice.** Pick what fits the task — most working analysts use both.
