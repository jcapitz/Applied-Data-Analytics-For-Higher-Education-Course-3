# 1.3 Choosing Your AI Coding Tool
## Codex vs Google Colab + Gemini
- You've now seen both tools
- Both support vibecoding — different approaches
- This lesson helps you choose (or use both!)

## How They Work: Key Difference

**Codex** = Separate AI agent running code in its own sandbox
```
YOU ──prompt──► CODEX ──results──► YOU
(You may never see the code)
```

**Colab + Gemini** = AI assistant inside YOUR notebook
```
┌── YOUR NOTEBOOK ──────────────┐
│  Your Code  ◄──► Gemini Help  │
│  (you control every line)     │
└───────────────────────────────┘
```

## Feature Comparison

| Feature | Codex | Colab + Gemini |
|:--------|:------|:---------------|
| Interface | Chat | Notebook + sidebar |
| Code visibility | Hidden | Fully visible |
| Iteration | Natural language | Edit code directly |
| GPU | No | Yes (free T4) |
| Collaboration | Share link | Real-time co-editing |
| Export | Chat only | .ipynb file |

## Cost Comparison

| Tier | Codex | Colab + Gemini |
|:-----|:------|:---------------|
| Free | Limited | Full Colab + basic Gemini |
| Paid | $20/mo (Plus) | $12/mo (Pro) |
| University | Enterprise | Google Workspace (often free) |

> For students: **Colab's free tier is more generous**

## Codex: Pros & Cons

### Advantages
- Lowest barrier — describe in English, get results
- End-to-end automation
- Great for quick exploration and prototyping

### Disadvantages
- Black box — code runs behind the scenes
- Less control over specific lines
- No GPU, no persistent workspace
- Requires subscription

## Colab + Gemini: Pros & Cons

### Advantages
- Full code control — see and edit every line
- Reproducible .ipynb notebooks
- Free GPU/TPU
- Google Drive integration
- **Learning-friendly** — seeing code helps you learn

### Disadvantages
- Requires some code literacy
- More manual work (cell by cell)
- Session timeouts (~90 min idle)

## When to Use Which?

```
Need to LEARN how the code works?
  → Colab + Gemini

Need a QUICK prototype?
  → Codex

Need GPU or team collaboration?
  → Colab + Gemini

Just exploring?
  → Either works!
```

## Scenario Guide for IR

| Scenario | Best Tool |
|:---------|:---------|
| Homework / coursework | Colab + Gemini |
| Quick data exploration | Codex |
| Formal report | Colab + Gemini |
| Prototyping new model | Codex |
| Team collaboration | Colab + Gemini |
| Learning new algorithm | Colab + Gemini |

## The Hybrid Workflow: Use Both!

### Pattern 1: "Codex First, Colab Second"
1. **Explore** with Codex — quick prototype
2. **Build** in Colab — proper notebook
3. **Share** the Colab notebook

### Pattern 2: "Colab Primary, Codex for Help"
1. **Work** in Colab as primary environment
2. **Stuck?** Ask Codex to diagnose
3. **Copy** the fix back to Colab

## Data Privacy Reminder

- Check your institution's data governance policies
- Both tools process data on cloud servers
- FERPA compliance requires institutional review
- **For this course:** use provided de-identified data

## The Bottom Line

| | Codex | Colab + Gemini |
|:--|:------|:---------------|
| Best for | Speed, prototyping | Learning, reproducibility |
| Code control | Low | High |
| Cost | $20/mo | Free |
| GPU | No | Yes |

> **There is no wrong choice.** Pick what fits your workflow, or use both.
