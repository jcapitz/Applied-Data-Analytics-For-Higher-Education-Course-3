# 1.3 Code Brief: Choosing Your AI Coding Tool

Quick reference for Codex vs Colab + Gemini.

## Quick Decision

| Scenario | Use |
|:---------|:----|
| Single notebook, one cloud file | **Colab + Gemini** |
| Homework / coursework | **Colab + Gemini** |
| Multi-file project with git history | **Codex** |
| Refactoring an existing repo | **Codex** |
| Writing tests next to model code | **Codex** |
| Learning a new algorithm cell-by-cell | **Colab + Gemini** |
| Real-time pair work | **Colab + Gemini** |
| Async PR-style review | **Codex** |
| Need free GPU | **Colab + Gemini** |
| No local Python / Chromebook | **Colab + Gemini** |

## Core Distinction

> **Codex** = CLI agent in your terminal that edits files across your repo (same family as Claude Code).
> **Colab + Gemini** = AI assistant inside one cloud notebook.

## Hybrid Workflow

```
1. EXPLORE in Colab    →  Sketch the approach in one notebook
2. PROMOTE with Codex  →  Turn it into a real repo with scripts, modules, tests
3. SHARE via git       →  Reproducible by anyone with the data + Python env
```

## Same CRISP Framework for Both

```
C — Context:      Describe your data
R — Role:         Specify your analytical perspective
I — Instructions: State the task clearly
S — Specifics:    Hyperparameters, features, methods
P — Product:      Desired outputs (charts, metrics, files)
```

## Privacy Reminder

Both tools send file/cell contents to a cloud model. Codex *running* locally does not mean your data *stays* local. Check institutional governance for both.
