# 1.2 Code Brief: AI-Assisted Coding with Codex

Condensed reference for using Codex effectively.

## Codex in 30 Seconds

- Install: `npm install -g @openai/codex` (or `brew install codex`)
- Auth: ChatGPT Plus/Pro/Team/Enterprise account, or API key
- Use: `cd <project>` → `codex` → describe what you want
- Codex reads/edits files in your repo and runs shell commands (with your approval)
- Conceptually equivalent to Claude Code

## CRISP Prompting Framework

```
C — Context:      "We have data/student_academics_data.csv with..."
R — Role:         "Act as an institutional research analyst..."
I — Instructions: "Create models/logreg.py that builds..."
S — Specifics:    "L2 penalty C=0.1, StandardScaler, 80/20 split, random_state=42"
P — Product:      "Save metrics to reports/logreg_metrics.json and ROC to reports/logreg_roc.png"
```

## Quick Prompt Templates

**EDA:**
```
Read data/<file>.csv. Create notebooks/eda.ipynb that shows summary
statistics, distributions, correlation heatmap, and the top features
correlated with [TARGET]. Run the notebook end-to-end.
```

**Classification:**
```
Create models/<name>.py that loads data/<file>.csv, fits a [MODEL]
with [PARAMS] on a StandardScaler-transformed 80/20 split, and writes
AUC/F1/precision/recall to reports/<name>_metrics.json plus a ROC
chart to reports/<name>_roc.png. Then run it.
```

**Comparison:**
```
Add reports/compare.py that loads metrics from each model's JSON,
prints an AUC/F1/Precision/Recall table, and saves an overlay ROC
plot to reports/compare_roc.png.
```

## Safety Defaults

- Work on a feature branch, not `main`
- Review every diff before approving
- Commit after each clean turn so you can `git restore` later
- Never approve commands you don't understand
