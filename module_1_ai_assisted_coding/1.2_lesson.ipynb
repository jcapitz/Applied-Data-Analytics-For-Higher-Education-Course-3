# 1.2 AI-Assisted Coding with Codex
- OpenAI's AI coding agent inside ChatGPT
- Writes, runs, and debugs Python in a cloud sandbox
- You describe what you want; Codex does the coding

## What is Codex?
- AI coding **agent** built into ChatGPT
- Writes and runs Python code autonomously
- Installs packages, reads your uploaded data
- Generates complete analyses from natural language

## What is Vibecoding?
Using natural language to direct AI to write code:
1. **Describe** what you want in plain English
2. **Review** the generated code
3. **Iterate** by refining prompts
4. **Validate** results with domain knowledge

> *You bring the institutional knowledge, Codex brings the coding speed.*

## Getting Started with Codex
1. Go to [chatgpt.com](https://chatgpt.com)
2. Requires ChatGPT Plus/Pro/Team/Enterprise
3. Select **Codex** in the model selector
4. Upload your data file and start prompting

## The Codex Workflow
```
1. Upload data file (CSV, Excel, etc.)
2. Write a prompt describing your analysis
3. Codex writes and runs Python code
4. Review output (tables, charts, metrics)
5. Refine prompt if needed
6. Download results
```

## The CRISP Prompting Framework
| Letter | Meaning | Example |
|:-------|:--------|:--------|
| **C** | Context | "I have 5,000 college students..." |
| **R** | Role | "Act as an IR analyst..." |
| **I** | Instructions | "Build a logistic regression to predict..." |
| **S** | Specifics | "L2 regularization, C=0.1, scale features..." |
| **P** | Product | "Show AUC, confusion matrix, top 5 coefficients" |

## Example: EDA Prompt
```
I've uploaded student records with HS_GPA, UNITS_ATTEMPTED,
GPA_1, DFW_RATE_1, and DEPARTED (0/1). Please:
1. Show summary statistics
2. Plot distributions
3. Show correlation heatmap
4. Top 3 features correlated with DEPARTED
```

## Example: Model Building Prompt
```
Build a regularized logistic regression:
- Target: DEPARTED
- Features: all numeric columns except DEPARTED
- StandardScaler, L2 with C=0.1
- Show: AUC, F1, precision, recall, confusion matrix
- Plot ROC curve
- Top 10 features by absolute coefficient
```

## Best Practices
### DO
- Be specific about metrics, charts, outputs
- Upload your actual data
- Iterate — first prompt rarely perfect
- Validate with domain knowledge

### DON'T
- Trust blindly — review the code
- Use vague prompts ("analyze my data")
- Share sensitive student data without governance approval

## The 80/20 Rule of Vibecoding
> Codex gets you **80%** of the way in **20%** of the time.
> Your job is the remaining 20%: **validation, interpretation, and institutional context.**
