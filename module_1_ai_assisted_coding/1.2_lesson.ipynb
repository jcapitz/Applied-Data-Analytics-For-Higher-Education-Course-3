# 1.2 AI-Assisted Coding with Codex
- OpenAI's AI coding **agent** that runs in your **terminal**
- Reads, edits, and runs files in your real project
- You describe what you want; Codex proposes diffs and commands you approve

## What is Codex?
- A command-line **agent** from OpenAI — same family as **Claude Code**
- Runs in your terminal, **not** in a chat window
- Reads/edits files across your **whole project**
- Runs shell commands (with your approval)
- Uses your real Python environment, your real data, your real git repo

## What is Vibecoding?
Using natural language to direct AI to write code:
1. **Describe** what you want in plain English
2. **Review** the generated code
3. **Iterate** by refining prompts
4. **Validate** results with domain knowledge

> *You bring the institutional knowledge, Codex brings the coding speed.*

## Getting Started with Codex
1. Install: `npm install -g @openai/codex` (or `brew install codex`)
2. Sign in with ChatGPT Plus/Pro/Team/Enterprise (or paste an API key)
3. `cd` into your project directory
4. Run `codex` — you're at the prompt, ready to give instructions

## The Codex Workflow
```
1. cd into your project, run `codex`
2. Describe a task in natural language
3. Codex reads relevant files, proposes diffs and commands
4. You review each diff / command, approve or edit
5. Codex applies edits, runs commands, reads output
6. Iterate via follow-ups — full project context persists
7. Commit the changes with git
```

## The CRISP Prompting Framework
| Letter | Meaning | Example |
|:-------|:--------|:--------|
| **C** | Context | "I have 5,000 college students..." |
| **R** | Role | "Act as an IR analyst..." |
| **I** | Instructions | "Build a logistic regression to predict..." |
| **S** | Specifics | "L2 regularization, C=0.1, scale features..." |
| **P** | Product | "Show AUC, confusion matrix, top 5 coefficients" |

## Example: Model Building Prompt
```
Create models/logreg.py that:

- Target: DEPARTED; features: all numerics except DEPARTED

```

## Best Practices
### DO
- Work in a **git repo** — git diff / restore is your safety net
- Review every **diff** before approving
- Name specific files, paths, and outputs in your prompts
- **Commit often** so you can roll back

### DON'T
- Don't run full-auto on `main` — use a branch
- Don't approve commands you don't understand (`rm -rf`, `git push --force`...)
- Don't share sensitive student data without governance approval — file *contents* go to OpenAI

## The 80/20 Rule of Vibecoding
> Codex gets you **80%** of the way in **20%** of the time.
> Your job is the remaining 20%: **validation, interpretation, and institutional context.**
