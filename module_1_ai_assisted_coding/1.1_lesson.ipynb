# 1.1 AI-Assisted Coding in Google Colab with Gemini

- Google Colab = free cloud-based Jupyter notebook
- Gemini = Google's AI assistant built into Colab
- No setup needed — runs in your browser

## What is Google Colab?

- Free, cloud-based Jupyter notebook by Google
- Zero setup — runs in browser
- Free GPU/TPU access
- Google Drive integration
- Pre-installed data science libraries

## What is Gemini in Colab?

Gemini is Google's AI assistant **inside** the notebook:

- Generate code cells from natural language
- Explain existing code
- Debug errors automatically
- Autocomplete as you type
- Chat in a side panel

## Why Colab + Gemini?

1. **Free** — no subscription needed
2. **Familiar** — Jupyter notebook interface
3. **Embedded AI** — no context switching
4. **Shareable** — like Google Docs
5. **University-friendly** — Google Workspace

## Getting Started

1. Go to [colab.research.google.com](https://colab.research.google.com)
2. Sign in with Google account
3. Click **"New Notebook"**
4. Click the **Gemini icon** (sparkle) in toolbar

## Three Ways to Use Gemini

1. **Chat Panel** — Click sparkle icon, type natural language
2. **In-cell generation** — Start with a comment, Gemini autocompletes
3. **Explain & Debug** — Right-click → "Explain" or click "Fix with Gemini"

## The CRISP Prompting Framework

| Letter | Meaning | Example |
|:-------|:--------|:--------|
| **C** | Context | "I have training.csv with student records..." |
| **R** | Role | "Help me as an IR analyst..." |
| **I** | Instructions | "Build a Random Forest to predict departure..." |
| **S** | Specifics | "Use 200 trees, StandardScaler..." |
| **P** | Product | "Show classification report, ROC curve, feature importance" |

## Colab-Specific Prompting Tips

- Reference cell outputs: "The dataframe from cell 3..."
- Ask for comments: "Generate code with comments explaining each step"
- Request visualizations: Gemini generates plots directly
- Chain cells: "Take the model from previous cell and plot SHAP values"
- Error context: Gemini sees error tracebacks automatically

## Hands-On: The Vibecoding Workflow

### prompt → review → iterate → validate

## Step 1: Setup & Data Loading

Tell Gemini:

> "Import pandas, numpy, sklearn. Upload training.csv. Show first 5 rows and shape."

## Step 2: Feature Engineering

Tell Gemini:

> "Create binary DEPARTED column (1 if SEM_3_STATUS != 'E'). Show class distribution."

## Step 3: Model Building

Tell Gemini:

> "Build logistic regression with L2, C=0.1 using HS_GPA, GPA_1, DFW_RATE_1. Scale features. Show AUC, F1, precision, recall."

## Step 4: Visualization

Tell Gemini:

> "Plot ROC curve. Create bar chart of feature coefficients."
