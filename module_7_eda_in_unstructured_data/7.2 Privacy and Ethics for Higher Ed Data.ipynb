## 1. Why Ethics Is Not Optional in Higher Ed Analytics


Ethical considerations in data science are often treated as an "add-on," but in Higher Education, they are foundational. As analysts, we must recognize that the data we use—often referred to as **administrative data**—was rarely collected with research or machine learning in mind.

### 1.1 Administrative Data Is Not “Free Data”
Grounding our work in **Tractenberg’s framing** (pages 3–4), we must distinguish between the original purpose of data collection and its secondary use for analytics. Administrative data is collected primarily to administer programs and support students, not to generate evidence for models.

**This creates a natural tension in our work:**
* **Statistical purposes vs. Non-statistical purposes:** Understanding the legal distinction (44 USC §3561) between analyzing patterns for policy and using data for direct administrative action.
* **The Promise of Confidentiality:** Maintaining the trust students place in the institution when they provide their information.
* **Data reuse tension:** Navigating the ethical boundaries of using data for a purpose other than why it was originally gathered.

> **Key Higher Ed Translation:**
> * **Student retention data** was collected for enrollment administration.
> * **LMS data** was collected for course delivery.
> * **Financial aid data** was collected for compliance.

> Using these sources for **predictive risk scoring** introduces entirely new ethical dimensions that require a deliberate approach.

### 1.2 The Expanding Dilemma
Using the "**Administrative Data Dilemmas**" framework (page 9), we must navigate several tensions that arise specifically in the higher ed context:
* **Secondary use** of data without explicit or fresh consent from the student.
* The balance between **Evidence-informed policy** vs. **Student autonomy**.
* The trade-off between **Fast, available data** vs. **Public trust**.
* Balancing **Institutional efficiency** vs. **Individual rights**.

## 2. Ethical Frameworks That Apply to You



This section anchors the legitimacy of our work. As data practitioners in higher education, we are guided by established professional codes that ensure our methods are both rigorous and responsible.



### 2.1 ASA Ethical Guidelines (2022)
The **American Statistical Association (ASA)** provides a comprehensive framework for ethical statistical practice. It is important to understand that "statistical practice" is defined broadly by the ASA:

> **Definition:** Statistical practice includes "designing the collection, summarizing, processing, analyzing, interpreting, presenting data; as well as model or algorithm development and deployment."

**What this means for you:** Your Machine Learning models—from the initial data split to the final deployment—fall directly under the **ASA Ethical Guidelines**.

**The 8 Major Principle Domains:**
* **A. Professional Integrity:** Maintaining high standards of competence and objectivity.
* **B. Integrity of Data and Methods:** Ensuring transparency and validity in how models are built.
* **C. Responsibilities to Stakeholders:** Being honest about the limitations and risks of your analysis.
* **D. Responsibilities to Data Subjects:** Protecting the privacy and dignity of the students whose data you use.
* **E. Multidisciplinary Teams:** Promoting ethical standards when working with IT, faculty, and leadership.
* **F. Profession:** Contributing to the collective improvement of the field.
* **G. Leadership:** Ensuring that organizational culture supports ethical practice.
* **H. Misconduct:** Identifying and reporting fraudulent or unethical behavior.

<br>

> **Instructor Emphasis:** While all domains matter, we will focus heavily on **Domain C (Stakeholders)**, **Domain D (Data Subjects)**, and **Domain B (Integrity of Methods)**. Additionally, the **Appendix** regarding organizational responsibility is a critical resource for those leading analytics teams.





### 2.2 ACM Code of Ethics
The **Association for Computing Machinery (ACM)** Code of Ethics is equally vital, especially as our work moves into complex computing and automated systems.

**Key Highlights from the ACM Code:**
* **Public good** is the primary consideration in all computing work.
* **Avoid harm** by proactively identifying and mitigating negative consequences.
* **Respect privacy** by ensuring students have control over their personal information.
* **Honor confidentiality** to protect sensitive institutional and student data.
* **Transparency** regarding how systems work and how decisions are made.
* **Robust and secure systems** that are technically sound and protected from misuse.
* **Leadership responsibility** in modeling ethical behavior for the institution.

<br>

* **Analytics directors** and Provosts

## 3. Ethical Practice Has Two Dimensions


Ethical practice in higher education analytics is not a static set of rules; it is an active process. According to Tractenberg (page 7), professional ethics requires navigating two distinct but related dimensions.




### 3.1 Dimension 1: Practice Ethically
This dimension is about the "how" of your daily work. It involves executing every technical task—from data cleaning to model deployment—according to established ethical standards.

**Examples include:**
* Ensuring **transparency** in how you handle missing data.
* Choosing models that prioritize **fairness** over marginal gains in accuracy.
* Documenting your process so it can be audited or peer-reviewed.

<br>



### 3.2 Dimension 2: Identify and Respond to Unethical Actions
This dimension is often much more difficult because it involves **interpersonal and institutional courage**. It requires you to recognize when a request or a process violates ethical standards and to take action.

**This often involves navigating "pushback" in scenarios such as:**
* Being asked to **remove a variable** because it makes a certain program look bad.
* Being pressured to **"make the numbers look better"** to meet a specific institutional goal.
* Identifying when risk scores are being used for **punitive decisions** rather than for student support.

> **Instructor Perspective:** Dimension 1 is about your technical integrity, while Dimension 2 is about your **professional responsibility** to the institution and the students. Identifying an unethical action is the first step; responding to it is what defines an ethical leader in data science.

## 4. Ethics Across the ML Pipeline


Ethics is not just a concern during data collection; it is a thread that must run through every phase of the machine learning lifecycle. As a professional steward of student data, you must evaluate ethical risks from the initial specification of needs to final model monitoring.

**The ethical lens applies to every stage of the process:**
* **Specification of needs:** Defining the problem without bias.
* **Data collection & Wrangling:** Ensuring representativeness and quality.
* **Modeling & Interpretation:** Avoiding "black-box" decisions that lack transparency.
* **Reporting & Deployment:** Communicating results honestly and responsibly.
* **Monitoring:** Tracking for model drift or emerging inequities over time.

<br>

**Applying Ethics to Your Course 2 Work:**
When you built your models in the previous course, every technical choice had an ethical dimension. For example, **Data splitting** involves the ethics of data leakage, **Imputation** can introduce hidden biases, **Class balancing** (like SMOTE) relates to representation ethics, and **Model deployment** carries significant downstream consequences for students.

| ML Stage | Ethical Questions for the Analyst |
| :--- | :--- |
| **Feature selection** | Are we encoding structural inequities? |
| **Scaling** | Are we masking subgroup patterns? |
| **SMOTE** | Are we synthetically altering real-world imbalance? |
| **Evaluation** | Are we optimizing for institutional gain or student well-being? |
| **Deployment** | Who is harmed if the model is wrong? |

> **Key Takeaway:** Ethical data science is a series of deliberate technical choices. By using this crosswalk, you ensure that potential harms are evaluated at every milestone of the development process.


## 5. Stakeholder Analysis Framework


When evaluating the ethics of a machine learning project, we must look beyond the immediate technical goals and identify who is impacted by our models. A **Stakeholder Analysis** allows us to map the potential consequences of our work across the institution.

| Stakeholder | Potential Harm | Potential Benefit | Unknowns / Risks |
| :--- | :--- | :--- | :--- |
| **You (The Analyst)** | Technical debt, ethical stress | Professional growth, impact | Long-term career alignment |
| **Your Supervisor** | Resource loss, reputational risk | Efficiency gains, goal met | Changing institutional priorities |
| **Provost / Leadership** | Policy failure, legal risk | Improved retention, data-driven | External political pressure |
| **Students** | Privacy loss, punitive action | Better support, graduation | Algorithmic bias, labeling |
| **Faculty** | Loss of autonomy, surveillance | Better student insights | Accuracy of engagement metrics |
| **Public Trust** | Institutional skepticism | Perceived transparency | Future data-sharing norms |
| **Unknown Groups** | Systemic exclusion | Unexpected positive outcomes | Downstream "black box" effects |

<br>

#### Core Principles of Stakeholder Analysis
As you complete this matrix for your own projects, keep two foundational principles from the framework in mind:

1. **All harms are not equal:** A minor inconvenience for an administrator (e.g., a slower reporting cycle) does not carry the same weight as a significant harm to a student (e.g., losing a scholarship due to a model error).
2. **Harms and benefits are not exchangeable:** You cannot "offset" a harm to one group by providing a benefit to another. Each cell in the matrix must be evaluated on its own ethical merits.

## 6. Ethical Reasoning Framework


Ethical practice is not just about following rules; it is about having a repeatable process for making difficult decisions. As outlined in the framework (page 10), professional ethical reasoning involves a six-step process to **make, justify, and revisit** decisions.

#### The 6-Step Reasoning Process
This framework is what transforms a technical notebook into a powerful tool for institutional leadership.

1. **Identify Prerequisite Knowledge:** Gather the necessary context, including professional guidelines (ASA/ACM), relevant laws (FERPA/HEOA), and the stakeholder analysis we completed in Section 5.
2. **Identify the Decision Framework:** Determine which ethical principles apply most directly to the current situation.
3. **Recognize the Ethical Issue:** Clearly state the conflict. (e.g., "This model increases efficiency but potentially marginalizes a specific student sub-population.")
4. **Evaluate Alternatives:** Brainstorm different technical or policy paths. What happens if we change the features? What if we change the intervention strategy?
5. **Make & Justify the Decision:** Choose a path and document the ethical rationale. This justification is your primary defense against future harm or liability.
6. **Reflect and Document:** After implementation, revisit the results. Did the decision achieve the intended benefit without causing unforeseen harm?

<br>

> **Instructor Perspective:** This process ensures that your decisions are **defensible**. In a higher ed setting, you may be called upon by leadership or faculty to explain *why* a certain model was deployed. Having a documented 6-step reasoning process moves the conversation from "the black box said so" to "we made a deliberate, ethical choice."

## 7. Higher Ed Applied Case Studies


This section moves our discussion from theory to application. As a professional steward of student data, you must be able to identify ethical tensions in real-world scenarios. For each case study below, consider the **Stakeholder Analysis** and the **6-Step Reasoning Process** we established.

#### **Case Study 1**: Reusing Survey Data for Predictive Modeling
In this scenario, a group of students completed a detailed mental health and wellness survey. The data was originally collected by the Counseling Center for service planning and resource allocation. However, the Analytics team now wants to use these responses to predict student attrition risk.

**Ethical Tensions to Consider:**
* **Consent Scope:** Did the students' original consent cover predictive modeling for retention?
* **Re-identification Risk:** How do we protect the identity of students in such a sensitive dataset?
* **Public Trust:** Will students stop being honest with counselors if they know the data is used for "scoring"?
* **Harm:** What are the consequences to the student if this sensitive data is breached?

<br>

#### **Case Study 2**: Predictive Risk Scores for Scholarship Allocation
An institution developed a high-performing model that predicts the probability of a student graduating on time. Leadership decides to use these "Graduation Probability Scores" to allocate a limited pool of emergency scholarship funds, prioritizing students with the highest scores to maximize the "return on investment."

**Ethical Tensions to Consider:**
* **Efficiency vs. Fairness:** Should funds go to those most likely to succeed, or those with the greatest need?
* **Self-Fulfilling Prophecy:** Does withholding funds from "low probability" students ensure they fail?
* **Structural Bias:** Does the model simply reward students who already have the most financial privilege?

<br>

#### **Case Study 3**: Using LMS Clickstream for Faculty Evaluation
The campus analytics team uses Learning Management System (LMS) clickstream data (e.g., login frequency, time spent on pages) to generate engagement metrics. While originally intended for student support, a new policy proposes using these metrics to rank instructor effectiveness during tenure and promotion reviews.

**Ethical Tensions to Consider:**
* **Data Repurposing:** Is it ethical to use data collected for student support to evaluate personnel?
* **Chilling Effects:** Will faculty change their teaching style just to "game" the engagement metrics?
* **Autonomy and Transparency:** Do instructors have a right to know exactly how they are being "scored" by an algorithm?

## 8. Documentation, Transparency, and Governance


Ethical practice is not complete without a commitment to transparency and formal governance. According to the framework (pages 13–14), documenting your choices is not just an administrative task—it is a critical safeguard for the institution and the students you serve.

**Key pillars of ethical governance include:**
* **Transparency in trade-offs:** Being open about the balance between model accuracy and other factors like fairness or student privacy.
* **Documenting alternatives:** Recording the other technical paths you considered and why they were rejected.
* **Justification as mitigation:** Providing a clear rationale for decisions to help mitigate potential harms and ensure institutional accountability.
* **Reflection-informed design:** Using the results of past models to inform the ethical design of future projects.

<br>

#### Institutional Governance Mechanisms
To move from individual ethical practice to institutional leadership, we must embed these values into our organizational structures.

**This involves utilizing and advocating for:**
* **Model Cards & Data Documentation:** Creating "nutrition labels" for every model that explain its purpose, training data, and known limitations.
* **Institutional Review Processes:** Ensuring that machine learning projects undergo ethical review similar to traditional human-subjects research.
* **Data Governance Boards:** Participating in cross-functional committees that oversee how student data is reused and protected.
* **Reflection Cycles:** Establishing formal periods to revisit deployed models and check for unintended consequences or "algorithmic drift."

> **Instructor Perspective:** Documentation is your primary tool for **transparency**. In higher education, where public trust is paramount, being able to provide a clear "paper trail" of your ethical reasoning is just as important as the accuracy of the model itself.
