# SECTION 1 — Why Ethics Is Not Optional in Higher Ed Analytics

### 1.1 Administrative Data Is Not “Free Data”

Ground in Tractenberg’s framing (page 3–4) :

* Administrative data is collected to administer programs — not to generate evidence.
* Statistical purposes vs non-statistical purposes (44 USC §3561).
* Promise of confidentiality.
* Data reuse tension.

**Key Higher Ed Translation:**

* Student retention data was collected for enrollment administration.
* LMS data was collected for course delivery.
* Financial aid data was collected for compliance.
* Using it for predictive risk scoring introduces new ethical dimensions.

# SECTION 3 — Ethical Practice Has Two Dimensions

From slide “Ethical Practice is more than ‘ethical practice’” (page 7) :

There are two dimensions:

### Dimension 1: Practice Ethically

Execute each task according to ethical standards.

### Dimension 2: Identify and Respond to Unethical Actions

Harder.
Often involves pushback.

Tie this to:

* Being asked to remove a variable.
* Being asked to “make the numbers look better.”
* Being pressured to use risk scores for punitive decisions.

# SECTION 6 — Ethical Reasoning Framework

From “Ethical Reasoning: Make, Justify, and Revisit Decisions” (page 10)

Teach the 6-step reasoning process:

1. Identify prerequisite knowledge (guidelines + laws + stakeholders)
2. Identify decision framework
3. Recognize ethical issue
4. Evaluate alternatives
5. Make & justify decision
6. Reflect and document
