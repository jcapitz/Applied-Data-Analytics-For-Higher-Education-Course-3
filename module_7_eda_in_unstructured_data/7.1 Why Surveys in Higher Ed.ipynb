# 7.1 Why Surveys in Higher Ed



## Introduction
Surveys play a foundational role in higher education analytics. While administrative systems capture structured data such as `GPA`, `enrollment status`, and `credit accumulation`, surveys capture something equally important: **student voice**.

Surveys allow institutions to measure:
* **Engagement** and **Belonging**
* **Satisfaction** and **Perceived Learning**
* **Academic Challenge** and **Support Experiences**

These are constructs that rarely appear in transactional databases but are central to **student success** and **institutional effectiveness**. In this section, we explore why surveys matter, how they are designed, and how they fit into the broader machine learning workflow.


## 1. Student-Centered Data Strategies
Modern higher education increasingly emphasizes **student-centered decision-making**. Institutions aim to understand not just *outcomes* (retention, GPA), but the *experiences* that shape them.

Surveys provide insight into:
* **Academic engagement**
* **Sense of belonging**
* **Faculty interaction**
* **Support utilization**
* **Workload and stress**
* **Perceived growth**




> **Key Insight:** Two students may have identical administrative trajectories (same GPA and retention). However, survey data may reveal that one felt **supported**, while the other felt **isolated**. That difference is critical for intervention.

Student-centered strategies rely on combining:
1. **Administrative data** (What happened)
2. **Survey data** (How it felt and why)

## 2. Prominent National Surveys

Higher education institutions frequently participate in nationally recognized surveys to benchmark performance.

Common examples include:

- **NSSE (National Survey of Student Engagement)**

- **SERU (Student Experience in the Research University)**

- **CIRP (Cooperative Institutional Research Program)**

- **CCSSE (Community College Survey of Student Engagement)**

These surveys measure constructs such as:

- Collaborative learning

- Academic challenge

- Faculty interaction

- Campus climate

- Support services

- Educational gains

National surveys allow institutions to compare results against peer institutions. They also provide validated instruments with established reliability and research backing.

From a machine learning perspective, these surveys generate rich structured data that can be merged with institutional datasets to explore relationships between engagement and outcomes.

## 3. Survey Design and Data Collection

Good survey data begin with **thoughtful design**. Survey construction involves:

- Clear construct definition

- Well-phrased items

- Appropriate response scales

- Representative sampling

Most survey items use Likert-type scales, such as:

- `Never` → `Very Often`

- `Strongly Disagree` → `Strongly Agree`

- `Not at all` → `Very much`

Survey administration methods may include:

- Online distribution

- Email invitations

- Embedded course evaluations

- Targeted subpopulation surveys

> **Note on Data Quality:** Low response rates can introduce **bias**, especially if certain student groups are underrepresented.

For institutional research professionals, survey methodology is not just about collecting responses — it is about ensuring validity, reliability, and representativeness.

## 4. Intra-Institutional Surveys


In addition to national instruments, many institutions develop internal surveys tailored to specific needs.

**Examples include:**
* **First-year experience** surveys
* **Advising satisfaction** surveys
* **Program-level climate** surveys
* **Exit surveys** for graduating seniors
* **Stop-out or withdrawal** surveys

**Internal surveys allow institutions to:**
* Measure **targeted initiatives**
* Evaluate **pilot programs**
* Investigate **emerging issues**
* Respond to **local priorities**

> **Key Insight:** Unlike national surveys, internal instruments can be adjusted quickly to address timely concerns. However, they require careful design to ensure **measurement consistency**.

From a data perspective, these surveys often produce both:
* **Structured Likert-scale** items
* **Open-ended text** responses

These text responses are especially important for later sections of this module.

## 5. Survey Methodology

Survey methodology determines how human responses are transformed into analyzable datasets. In a machine learning pipeline, we must treat survey data with the same structural rigor as administrative data, organizing it into **Sampling Units** (Rows) and **Item Responses** (Columns).


### 5.1 Rows — Sampling Units

In most survey datasets, each row represents a single **respondent** (typically a unique student). This row-per-student structure is what allows us to merge survey insights with administrative data like `GPA` or `retention_status`.

> **Instructor Note:** When preparing data for modeling, each row becomes a single observation in your **feature matrix**. Because each row corresponds to a unique individual, we can use these observations to predict specific student outcomes.

**Example Row Structure:**
| Student_ID | Q1 | Q2 | Q3 | ... |
| :--- | :--- | :--- | :--- | :--- |
| `1001` | 4 | 5 | 2 | ... |
| `1002` | 2 | 3 | 5 | ... |

### 5.2 Columns — Item Responses

Each column represents a specific survey item or variable. These columns serve as the **Features** for our models. Depending on the question type, these columns contain different data formats:

* **Likert-scale responses:** Numeric-coded values (e.g., 1–5) representing levels of agreement or frequency.
* **Binary responses:** Simple `Yes/No` or `0/1` indicators.
* **Categorical demographic fields:** Descriptive tags (e.g., "Major" or "Campus").
* **Open-ended text responses:** Unstructured student comments.

**Example Feature Set:**
| Student_ID | Collaborative_Learning | Faculty_Interaction | Belonging_Score | Open_Response |
| :--- | :--- | :--- | :--- | :--- |
| `1001` | 4 | 3 | 5 | "I felt very supported..." |

<br>
<br>

| Feature Type | Processing Requirement |
| :--- | :--- |
| **Likert / Numeric** | Often used directly as ordinal data |
| **Categorical** | Requires **One-Hot Encoding** |
| **Text Responses** | Must be transformed into **Numerical Vectors** |

> **Key Takeaway:** Survey columns are unique because they provide both **structured** and **unstructured** data sources, requiring different preparation techniques before they are "model-ready."

## 6. The UPAD Cycle for Survey Data

Survey data follow the same **UPAD cycle** used throughout this course. However, because we are dealing with human perceptions and self-reported measures, each stage requires specific attention to detail to ensure the model remains accurate and representative.




### **Understand**
Before touching the data, you must define the **constructs** being measured. This involves identifying what each item represents (e.g., does "Q1" measure *Academic Challenge* or *Financial Stress*?) and understanding the response scales used, such as 1-5 Likert-type scales.
<br>

### **Prepare**
Preparation is often the most critical step for survey data. To make the data "model-ready," you must:
* **Clean missing responses:** Students frequently skip questions, requiring careful imputation or removal.
* **Encode Likert scales:** Convert text-based agreements into numeric values.
* **Handle non-response bias:** Ensure your sample represents the entire student body, not just those who chose to respond.
* **Transform text data:** If open-ended comments are present, they must be vectorized for analysis.
<br>

### **Analyze**
In the analysis phase, we move from raw responses to actionable features:
* **Explore distributions:** Visualize how student sentiments are spread across the population.
* **Assess correlations:** Determine how survey constructs relate to outcomes like GPA or retention.
* **Build predictive models:** Use the survey items as "Features" to identify students who may need support.
<br>

### **Deploy**
The ultimate goal of the cycle is institutional improvement:
* **Inform policy decisions:** Use aggregate feedback to change campus strategies.
* **Target interventions:** Reach out to specific students based on their reported needs.
* **Monitor trends:** Track how student perceptions shift over multiple semesters.

> **Key Insight:** Survey data are not separate from machine learning workflows. They are fully integrated into structured modeling pipelines once they are properly prepared and validated.

## 7. Conclusion


Surveys are a cornerstone of higher education analytics because they capture what administrative systems cannot: **student perception, engagement, and self-reported growth**. They provide the "why" behind the "what" found in institutional databases.

**In summary, survey data allows institutions to:**
* **Measure learning experiences** and academic challenge.
* **Identify equity gaps** through the student lens.
* **Detect dissatisfaction early**, before it leads to student departure.
* **Evaluate support systems** and campus initiatives.
* **Connect experience to outcomes** to drive institutional effectiveness.

> **Final Thought:** Understanding the structure and purpose of these instruments is the first step in turning student voice into actionable insight. Survey data are not merely descriptive tools — they are rich, analyzable inputs that can enhance predictive modeling and institutional decision-making.
